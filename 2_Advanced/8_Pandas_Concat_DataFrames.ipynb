```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

# **Concatenating January and February jobs**

```python
# Sample dataset of job postings in January
job_postings_jan = pd.DataFrame({
    'job_id': [1, 2, 3, 4, 5],
    'job_title': ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer', 'Data Scientist', 'Data Engineer'],
    'company': ['Company A', 'Company B', 'Company C', 'Company D', 'Company E'],
    'job_posted_date': pd.to_datetime(['2024-01-02', '2024-01-07', '2024-01-14', '2024-01-19', '2024-01-24'])
})

job_postings_jan
```

```python
job_postings_feb = pd.DataFrame({
    'job_id': [6, 7, 8, 9, 10],
    'job_title': ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer', 'Data Scientist', 'Data Engineer'],
    'company': ['Company F', 'Company G', 'Company H', 'Company I', 'Company J'],
    'job_posted_date': pd.to_datetime(['2024-02-05', '2024-02-09', '2024-02-12', '2024-02-18', '2024-02-22'])
})

job_postings_feb
```

```python
pd.concat([job_postings_jan, job_postings_feb])
```

```python
# The output above returns an index of 0-4 twice.
# To create new index, we ignore the current index

pd.concat([job_postings_jan, job_postings_feb], ignore_index=True)
```

# **In order to concatenate with our original data, we have to create a new data**

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')

df
```

```python
months = df['job_posted_month'].unique()

months
```

```python
# Creating a dictionary of all month values

{month: month for month in months}     # Recall: output of a dictionary do not come in any particular order or sort.
```

```python
# Rather than returning the month in our dictionary value field,
# Let's return the whole dataframe but filtered for posts matching month value in the job_posted_month column
# Basically where job_posted_month == months (variable assigned to unique job_posted_month value)
# Then assign variable - dict_months

dict_months = {month: df[df['job_posted_month'] == month] for month in months}
```

```python
# Outputing January job postings

dict_months['Jan']
```

# **2023 QUARTER 1**

```python
pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)
```

```python
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)
```

```python
df_q1['job_posted_month'].value_counts().plot(kind='bar')
plt.title('Job Postings in First Quarter, 2023')
plt.ylabel('Job Counts')
plt.xlabel('')
plt.show()  
```
