```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
```

# **Creating a new columnn of month name from the job_posted_date**

```python
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')

df_US
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

df_US_pivot
```

# Cleaning Up the dataframe in order to sort in chronological order

```python
df_US_pivot.reset_index(inplace=True)

df_US_pivot
```

```python
# Create a new column for month number of the equivalent job_posted_month to serve as our index
# With this, we can be able to sort our dataframe in chronological order

df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
```

```python
df_US_pivot
```

```python
# Now we have the month number (e.g Jan=1, April=4),
# we can now sort in chronological order
# and not alphabetical order (which will make no statistical sense).

df_US_pivot.sort_values('month_no', inplace=True)

```

```python
df_US_pivot
```

```python
# Setting the job posted month back to index

df_US_pivot.set_index(keys='job_posted_month', inplace=True)

```

```python
df_US_pivot
```

```python
# Dropping the month_no column from our pivot table,
# as we have no use for it in our final pivot table presentation.

df_US_pivot.drop(columns='month_no', inplace=True)
```

```python
df_US_pivot
```

```python
# Dropping the "index" column

df_US_pivot.drop(columns='index', inplace=True)
```

```python
df_US_pivot
```

```python
# Plotting the final pivot table

df_US_pivot.plot(kind='line')
```

```python
# Further Clean-up to view the top 3 counts of jobs posted:

df_US['job_title_short'].value_counts().head(3)
```

```python
# Assigning it a variable.
# Then turning it into a list, in order to use it as filter

top_3 = df_US['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
top_3
```

```python
# Plotting the chart

df_US_pivot[top_3].plot(kind='line')
```

```python
# Adding Labels

df_US_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in US')
plt.ylabel('Job counts')
plt.xlabel('2023')
plt.show()
```
