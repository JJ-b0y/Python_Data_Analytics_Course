```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
import ast

# Advanced visualization
import seaborn as sns

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

# Further Cleanup: job_Skills to list datatype
df['job_skills'] = df['job_skills'].apply(lambda y: ast.literal_eval(y) if pd.notna(y) else y)
```

```python
df_DA_US = df[(df['job_title'] == "Data Analyst") & (df['job_country'] =="United States")].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_US_pivot
```

```python
# Creating a new row of "Total"

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

# Filtering the dataframe by the newly created "Total" row

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_US_pivot
```

```python
# Deleting the "Total" row from our dataframe after establishing/sorting the order

df_DA_US_pivot = df_DA_US_pivot.drop('Total')

df_DA_US_pivot
```

# **What is the percentage of each skill appearing in Job Postings, relative to others?**

```python
# Converting the Dataframe from count display to percentage

df_DA_US.groupby('job_posted_month_no').size()

# Assigning a variable name and running it

DA_totals = df_DA_US.groupby('job_posted_month_no').size()

DA_totals
```

```python
# After getting the totals per month for each skill, we need to divide the appropriate row by the total to get the percentage

df_DA_US_percent = df_DA_US_pivot.div(DA_totals/100, axis=0)

df_DA_US_percent
```

```python
# Month no to Month name

df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')
```

```python
df_DA_US_percent
```

# **Plotting the Chart**

```python
df_plot = df_DA_US_percent.iloc[:, :5]      # All Rows, First 5 Columns

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()   # Removes border

plt.title('Trending Top Skills for Data Analyst in US')
plt.ylabel('Likelihood in Job Postings')
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter  # display y-axis in percentage
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(11.1, df_plot.iloc[-1, i], df_plot.columns[i])
```
