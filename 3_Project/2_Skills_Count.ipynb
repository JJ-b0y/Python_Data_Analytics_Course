```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
import ast

# Advanced visualization

import seaborn as sns

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

# Further Cleanup: job_Skills to list datatype
df['job_skills'] = df['job_skills'].apply(lambda y: ast.literal_eval(y) if pd.notna(y) else y)
```

```python
df[df['job_country'] == 'United States']
df_US = (df[df['job_country'] == 'United States'])
```

```python
df_US
```

```python
df_skills = df_US.copy()
df_skills = df_skills.explode('job_skills')
```

```python
df_skills
```

```python
skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size()  # Grouping the df by skills and job_title
```

```python
skills_count
```

```python

df_skills_count = skills_count.reset_index(name='skill_count')      # Giving our count column a name
df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)

df_skills_count
```

```python
job_titles = df_skills_count['job_title_short'].unique().tolist()  # Calling the unique values in job_title_short returning the outcome in a list format.

job_titles = sorted(job_titles[:3])     # Returns Top 3 job titles and sorted in alphabetial order.

job_titles

```

```python
fig, ax = plt.subplots(len(job_titles), 1)        # 3 rows and 1 column

for i, job_title in enumerate(job_titles):
    df_plot = (df_skills_count[df_skills_count['job_title_short'] == job_title].head(5))
    df_plot.plot(kind='barh', x='job_skills', y='skill_count', ax=ax[i], title=job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].legend().set_visible(False)
    # ax[i].set_xlim(0, 120000)

fig.suptitle('Counts of Top Skills in US Data Job Postings', fontsize=15)
fig.tight_layout()
```

# **We want to show the top skills above in percentage**

```python
df_US['job_title_short'].value_counts().reset_index(name='jobs_total')
```

```python
df_job_title_count = df_US['job_title_short'].value_counts().reset_index(name='jobs_total') # Assigning the previous df to a variable
```

```python
df_skills_perc = pd.merge(df_skills_count, df_job_title_count, how='left', on='job_title_short')

df_skills_perc

# Left Join df_skills_count on df_job_title_count, that is maintain the left column (skill_count) then fill in the corresponding values on the right (jobs_total)
# Assigning the merge a variable- df_skills_perc
```

```python
# To get the percentage of skill per job_title

df_skills_perc['skill_count'] / df_skills_perc['jobs_total']
```

```python
# To percent value, multiply by 100

100 * df_skills_perc['skill_count'] / df_skills_perc['jobs_total']
```

```python
# Add the percent column to our dataframe

df_skills_perc['skill_percent'] = 100 * df_skills_perc['skill_count'] / df_skills_perc['jobs_total']

df_skills_perc
```

```python
# Building on from our earlier graph to plot the skills percent

fig, ax = plt.subplots(len(job_titles), 1)        # 3 rows and 1 column

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_perc[df_skills_perc['job_title_short'] == job_title].head(5)
    df_plot.plot(kind='barh', x='job_skills', y='skill_percent', ax=ax[i], title=job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].legend().set_visible(False)
    ax[i].set_xlim(0, 78)   # So that all 3 charts will be on the same x-axis

fig.suptitle('Likelihood of Skills Requested in US Data Job Postings', fontsize=15)
fig.tight_layout(h_pad=0.5) # Fix overlap
plt.show()
```

# **Advanced Visualization Using Seaborn**

```python
fig, ax = plt.subplots(len(job_titles), 1)        # 3 rows and 1 column

sns.set_theme(style='ticks')

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_perc[df_skills_perc['job_title_short'] == job_title].head(5)
    # df_plot.plot(kind='barh', x='job_skills', y='skill_percent', ax=ax[i], title=job_title) <-- This is from our matplotlib, replace by seaborn
    sns.barplot(data=df_plot, x='skill_percent', y='job_skills', ax=ax[i], hue='skill_count', palette='dark:b_r')
    ax[i].set_title(job_title)
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].get_legend().remove()
    ax[i].set_xlim(0, 78)   # So that all 3 charts will be on the same x-axis

    # Displaying the percent value beside each bar
    for n, v in enumerate(df_plot['skill_percent']):
        ax[i].text(v + 1, n, f'{v:.0f}%', va='center')

    # Removing the x-axis of the top 2 charts, to display only one x-axis (one for all) at the bottom.
    if i != len(job_titles) - 1:
        ax[i].set_xticks([])

fig.suptitle('Likelihood of Skills Requested in US Data Job Postings', fontsize=15)
fig.tight_layout(h_pad=0.5) # Fix overlap
plt.show()
```
